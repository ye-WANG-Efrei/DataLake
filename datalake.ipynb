```
# create a Temporary dataframe
df_t=df
```

```
# AdjClose/open -1; convert float to .2%; add this new column
df_t['Daily_return_rate']= df.apply(lambda x: format((float(x["AdjClose"])/float(x["open"])-1), '.2%'), axis=1, result_type='expand')
```

```
#Moving average
def Movingaverage(df_t):
    # groupby company_name
    gpb=df_t.groupby("company_name") 
    #list(gpb):[name(string), stocks_info(dataframe)]
    #sort by date
    #sliding window each 7 rows as one window  as 7-Day Moving Average
    #add moving Constants = 5
    df_MA=pd.DataFrame()
    for i in range(len(list(gpb))):
        print("df_MA len : ",len(df_MA))
        df_MA_t=list(gpb)[i][1].sort_values(by = 'Date')
        df_MA_t['7-Day Moving Average'] = df_MA_t['open'].rolling(7).mean()+5
        df_MA = df_MA.append(df_MA_t)
    return df_MA
df = Movingaverage(df_t)
```
