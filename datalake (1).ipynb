```
#Moving average
def Movingaverage(df_t,timep):
    colname="DMA"+str(timep)
    # groupby company_name
    gpb=df_t.groupby("company_name") 
    #list(gpb):[name(string), stocks_info(dataframe)]
    #sort by date
    #sliding window each 7 rows as one window  as 7-Day Moving Average
    #add moving Constants = 5
    df_MA=pd.DataFrame()
    for i in range(len(list(gpb))):
        #print("df_MA len : ",len(df_MA))
        #de-duplication
        df_MA_t=list(gpb)[i][1].sort_values(by = 'Date').drop_duplicates()
        df_MA_t[colname] = df_MA_t['open'].rolling(timep).mean()+int(constant)
        df_MA = df_MA.append(df_MA_t)
    return df_MA
df = Movingaverage(df_t,7)
```

```
#Customization parameters
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'
}
crypto_to_scrape = ['BTC', 'ETH', 'ADA']
real_time_prices = {}
email_addreess = '' #<YourEmailAddress>
app_password = '' # <YourAppPassword>
```

```
#scrapy using requests and bs4
def scrape(url):
    r = requests.get(url, headers=headers)
    data = r.content
    soup = BeautifulSoup(data, 'lxml')
    return soup

def format_price(price):
    return float(''.join(price.split(',')))


def dailydata(crypto):
    dailydata = {}
    historical_url = f'https://sg.finance.yahoo.com/quote/{crypto}-EUR/history?p={crypto}-EUR'
    soup = scrape(historical_url)
    rows = soup.findAll('tr', attrs={'class': 'BdT Bdc($seperatorColor) Ta(end) Fz(s) Whs(nw)'})
    for i in range(0,len(rows)):
        date = rows[i].find('td', attrs={'class': 'Py(10px) Ta(start) Pend(10px)'}).text
        open_price = rows[i].find('td', attrs={'class': 'Py(10px) Pstart(10px)'}).text
        open_price = format_price(open_price)
        if i != 0:
            dailydata[date]=open_price
    #print('past_seven_day_open',past_seven_day_open)
    
    #print("Avg value of " + crypto + " on 7 days : %.2f" % t_sum)
    return dailydata


def find_real_time_price(crypto):
    yahoo_url = f"https://sg.finance.yahoo.com/quote/{crypto}-EUR?p={crypto}-EUR"
    soup = scrape(yahoo_url)
    price = soup.find('fin-streamer', attrs={'class': 'Fw(b) Fz(36px) Mb(-4px) D(ib)'})
    real_time_prices[crypto] = format_price(price.text)
    return format_price(price.text)
```

```
txt_dialy=""
for crypto in crypto_to_scrape:
    dailydic=dailydata(crypto)
    dailydicTOjson = json.dumps(dailydic)
    txt_dialy+= dailydicTOjson
for crypto in crypto_to_scrape:
    realtime_p=find_real_time_price(crypto)
    dic=dict(time=datetime.now().timestamp(),price = realtime_p)
    print(dic)
```
